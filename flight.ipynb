```python
# feature engineering

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
# date month and year here as object , we  have to convert it into integer
```

```python
# converting  object into integer
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
# drop date of journery
df.drop('Date_of_Journey',axis =1,inplace = True)
```

```python
df.head()
```

```python
# changing arivial time and split it into hr and minute 


# df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]

# df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]



# df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])

# df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]



df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0].split(':')[0])


df['Arrival_minute'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0].split(':')[1]) 

```

```python
# converting this object to integer

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)

```
